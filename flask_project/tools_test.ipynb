```python
with open("./stopwords.txt",encoding='utf-8')as s:
    content1=s.read()
stop_words=content1.split()
```

```python
seg_list = jieba.cut(text)
# 排除停用词并统计词频
filtered_words = [word for word in seg_list if word not in stop_words]
word_counts = Counter(filtered_words)
```
